```python
# Coding: utf-8

"""
Created On 19th Dec 2020
@author: Nikhil Jagnade
"""

import os
import time
import requests
import sys

def retrieve_html():
    for year in range(2013,2019):
        for month in range(1,13):
            if(month<10):
                url='https://en.tutiempo.net/climate/0{}-{}/ws-432950.html'.format(month, year)
            else:
                url='https://en.tutiempo.net/climate/{}-{}/ws-432950.html'.format(month, year)
        
            texts = requests.get(url)
            # print("text = ", texts)
            text_utf = texts.text.encode('utf_8')
            # print("text_utf", text_utf)
        
            if not os.path.exists("C:/Users/dell/Live Projects - DSAIML/AQI Prediction/Data/Html_Data/{}".format(year)):
                os.makedirs("C:/Users/dell/Live Projects - DSAIML/AQI Prediction/Data/Html_Data/{}".format(year))
            with open("C:/Users/dell/Live Projects - DSAIML/AQI Prediction/Data/Html_Data/{}/{}.html".format(year, month),"wb") as output:
                output.write(text_utf)
                #print("path=", output)
        
        sys.stdout.flush()
        
if __name__ == '__main__':
    start_time = time.time()
    retrieve_html()
    stop_time = time.time()
    print("Time taken {}".format(stop_time-start_time))           
```
